```python
import pandas as pd
import os
import json
import plotly.express as px
```

```python
folder_path = "data"

dataframes = []

for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, 'r') as file:
            data = json.load(file)['wines']

        if isinstance(data, list):
            for item in data:
                df = pd.json_normalize(item)
                dataframes.append(df)

        else:
            df = pd.json_normalize(data)
            dataframes.append(df)

final_dataframe = pd.concat(dataframes, ignore_index=True)
```

```python
to_drop = [
    "region.background_image", "style.background_image", 'region.background_image.location', 'region.background_image.variations.large', 'region.background_image.variations.medium',
    'region.country.currency.suffix', 'style.blurb', 'style.image', 'style', 'taste.structure',
    'style.vintage_mask', 'style.baseline_structure.fizziness', 'taste.structure.fizziness', 'statistics', 'winery.background_image'
]

final_dataframe.drop(to_drop, axis=1, inplace=True)
```
